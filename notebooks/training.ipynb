# Training Notebook

This notebook trains the model using the provided dataset.


```python
# Import necessary modules
import os
import torch
from ultralytics import YOLO
```

```python

def train(data_path, model_name, epochs, device):
    """Train a YOLO model with the given parameters."""
    model = YOLO(model_name)  # Load a YOLO model (e.g., 'yolov5s.pt')

    # Train the model
    model.train(
        data=data_path,  # Path to data.yaml
        epochs=epochs,
        device=device,  # '0' for GPU, 'cpu' for CPU
        imgsz=640,  # Image size
        batch=16,  # Batch size
    )

    # Save the trained model
    model.save(os.path.join("models", "yolo_trained_model.pt"))
```

```python
# Define training pipeline
if __name__ == "__main__":
    data_path = "data.yaml"  # Path to the dataset configuration
    model_name = "yolov11.pt"  # Pretrained YOLO model (Updated to YOLOv11)
    epochs = 10
    device = "cuda" if torch.cuda.is_available() else "cpu"
```
